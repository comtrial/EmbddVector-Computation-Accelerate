```python

import torch
from sentence_transformers import SentenceTransformer, util

model = SentenceTransformer("Huffon/sentence-klue-roberta-base")
```

```python
if torch.cuda.is_available():
    device = torch.device('cuda:0')
else:
    device = torch.device('cpu')
```

```python
model.to(device)
```

```python
import pandas as pd

df = pd.read_excel('/content/drive/MyDrive/colab_data/personal_project/aihub_or_kr-sports_ko.xlsx')
```

```python
df['embedding'] = df['text'].map(lambda x:model.encode(x).tolist())
df.to_pickle('/content/drive/MyDrive/colab_data/personal_project/embedding.pkl')
```
