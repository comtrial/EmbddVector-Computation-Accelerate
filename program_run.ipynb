## 10차원으로 축소한 모델

```python

def main():
    
    start = time.time()
    # DataLoader
    print('Data 로딩 중..') 
    EMBD_PICK_DIR = 'data/embedding.pkl'
    dataset = EmbeddDataset(EMBD_PICK_DIR)
    data_loader = DataLoader(dataset, batch_size=32, shuffle = False)

    # Autoencoder 
    print('차원축소 모델 설정 중..')
    target_dim = 10
    device = torch.device('cpu')
    model_path = "saved_model/AE10D.ckpt"
    model = Autoencoder(target_dim)
    model.load_state_dict(torch.load(model_path, map_location=device)["state_dict"])

    # 학습된 Autoencoder를 통한 차원축소된 20차원 vector
    print('차원축소 중..')
    encoded_vector = get_encodings(model, data_loader).numpy()

    # 메모리 해제    
    del dataset, data_loader
    


    # faiss 벡터 검색엔진
    print('검색 중..')
    engine = VectorEngine()
    df = pd.read_excel('/content/drive/MyDrive/colab_data/Embedding_Search/data/aihub_or_kr-sports_ko.xlsx')
    encoded_vector = np.random.rand(len(df), 10).astype('float32')
    index, db_vector = engine.indexer(encoded_vector)
    result = engine.searcher(df, db_vector, index)
    
    result.sort_values(by = 'similarity', ascending=False).to_csv('result.csv')
    
    print(f'프로그램 실행 시간 {time.time() - start}초, 찾은 결과{len(result.shape)}')
    
    return

main()
```
